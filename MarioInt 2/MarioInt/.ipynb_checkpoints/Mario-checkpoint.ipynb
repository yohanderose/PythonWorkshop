```python
# Import some libraries that use in this notebook
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import numpy as np

from IPython.display import clear_output

# import the widgets for silder movement
from ipywidgets import interact, interactive, fixed, interact_manual

# Import the images that are used. Note that they have to be located in
# same folder as the notebook or a directory has to be specified
img=mpimg.imread('background.png')
img2=mpimg.imread('mario.png')
img3=mpimg.imread('coin.png')
img4=mpimg.imread('mario_jump.png')
```

```python
# function to plot mario. Do not change this!
def plot_mario(mario_x= 0, mario_y= 2, coin_app=True):
    clear_output(wait=True)
    
    c_positions = np.array([[15,6]])
    
    coin = c_positions
    plt.figure(figsize=(18,10))
    
    plt.imshow(img, extent=(0,20,0,12))
    if mario_y == 2:
        plt.imshow(img2, extent=(mario_x,mario_x+2, mario_y, mario_y+2))
    else:
        plt.imshow(img4, extent=(mario_x,mario_x+2, mario_y, mario_y+2))
    for i in c_positions:
        if coin_app is True:
            plt.imshow(img3, extent=(i[0],i[0]+2, i[1], i[1]+2))
        else:
            pass
    plt.xlim(0,20)
    plt.ylim(0,10)
    plt.minorticks_on()
    plt.grid(which='minor', linestyle=':', linewidth=0.2, color='black')
    plt.grid()
```

```python
# for loop replaced by function that can be called from interact widget
def move_mario(x_coord):
    coin = True
    if x_coord == 15:
        coin=False
    if x_coord>=6 and x_coord<17:
        plot_mario(x_coord, (-0.2*(x_coord-11)**2)+8, coin)
    
    else:
        plot_mario(x_coord,2)
```

```python
interact(move_mario, x_coord = (0,20,1))
```
